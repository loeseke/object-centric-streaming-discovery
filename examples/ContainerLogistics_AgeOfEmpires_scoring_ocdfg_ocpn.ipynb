```python
import sys
sys.path.append('../src')
from cache_policy_buffers import CachePolicy
from vars import *
from onl_vs_offl_evaluation_ocdfg_ocpn import plot_avg_score_over_buf_sizes, plot_heatmap_cp_x_pp

from pathlib import Path
```

### Plotting evaluation scores for varying model-buffer sizes

```python
# Visualize progression of average evaluation scores for fixed cache and priority policy and varying buffer sizes
# Note that, per run, the model buffers are all set to the same size
for model in ['ocdfg', 'ocpn']:
    for log_name, buf_sizes in [('ContainerLogistics', [5] + list(range(50, 1150, 50))),
                                ('AgeOfEmpires10Matches', [5] + list(range(250, 1500, 250)))
                                ]:
        plot_avg_score_over_buf_sizes(
            file_path=f'../data/{log_name}.json',
            buf_sizes=buf_sizes,
            model_buf_name=model,
            cp=CachePolicy.FIFO,
            pp_buf=None,
            coupled_removal=False,
            output_dir=Path(f'../figures/{log_name}/{model}_scoring')
        )
```

### Plotting average evaluation scores for all combinations of cache and priority policy

```python
# Visualize accuracy, precision, and recall heatmaps for all cache policy and type-based priority policy combinations and with and without coupled removal
# Note that parameters of PPs are fixed
for log_name, buf_size in [('ContainerLogistics', 125), 
                           ('AgeOfEmpires10Matches', 250)
                          ]:
    print(f'### Running OC-DFG on {log_name}...')
    plot_heatmap_cp_x_pp(
        file_path=f'../data/{log_name}.json', 
        model_buf_name='ocdfg',
        buf_size=buf_size,
        output_dir=Path(f'../figures/{log_name}/ocdfg_scoring')
    )
```
